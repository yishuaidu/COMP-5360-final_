##### Las Vegas Crawler

Data about each Zip code from Las Vegas is scraped from website https://statisticalatlas.com/metro-area/Nevada/Las-Vegas/Overview. The dataset file called df_zipcode_clean.csv. We also combine two datasets together.

Note: the code of Las Vegas Crawler, we have another file called Las_Vegas_Crawler.ipython

```python

### normalize
from transform import transform_number as tn  ## this is the transform module that we built

df_zip = pd.read_csv("df_zipcode_clean.csv")
df_zip = df_zip.drop('Unnamed: 0', 1)
df_zip = df_zip.drop('Unnamed: 0.1', 1)
df_zip["rate_high_edu"] = tn.normalize(df_zip['edu_high'],df_zip['Population'])
df_zip["rate_Emp"] = tn.normalize(df_zip['Emp'],df_zip['Population'])
df_zip["rate_Race_w"] = tn.normalize(df_zip['Race_w'],df_zip['Population'])
df_zip["rate_Race_h"] = tn.normalize(df_zip['Race_h'],df_zip['Population'])
df_zip["rate_Race_b"] = tn.normalize(df_zip['Race_b'],df_zip['Population'])
df_zip["rate_Race_a"] = tn.normalize(df_zip['Race_a'],df_zip['Population'])
df_zip["rate_Race_m"] = tn.normalize(df_zip['Race_m'],df_zip['Population'])
df_zip["rate_Race_o"] = tn.normalize(df_zip['Race_o'],df_zip['Population'])
df_zip['rate_business'] = [len(data[data.postal_code == i])/len(data) for i in df_zip.zip.tolist()] 

df_zip = df_zip[df_zip.rate_business > 0.015658]
zip__ = df_zip.zip
df_zip = df_zip[['rate_Race_w','rate_Race_h','rate_Race_b','rate_Race_a','rate_Race_m','rate_high_edu','rate_Emp','income','rate_business']]

df_zip.head()


df_zip.describe()
```

##### Solve problem by Clustering(PCA and Kmeans)

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering


pd.set_option('display.notebook_repr_html', False)

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
plt.style.use('ggplot')

## save in array for using scale() function 
df2 = df_zip[['income','rate_high_edu','rate_Emp','rate_Race_w','rate_Race_h','rate_Race_b','rate_Race_a','rate_Race_m']]

df3=np.array(df2)

# scale the dataset
data1 = scale(df3[:])

```

```python
pca_model = PCA()

PCA = pca_model.fit_transform(data1)


# create a new pandas dataframe 
PC1 = []
PC2 = []
PC3 = []
PC4 = []
PC5 = []
PC6 = []
PC7 = []
PC8 = []




for i in PCA:
    PC1.append(i[0])
    PC2.append(i[1])
    PC3.append(i[2])
    PC4.append(i[3])
    PC5.append(i[4])
    PC6.append(i[5])
    PC7.append(i[6])
    PC8.append(i[7])



df3 = pd.DataFrame({ '' : zip__, "PC1" : PC1, "PC2" : PC2,"PC3" : PC3,"PC4" : PC4,"PC5" : PC5,"PC6" : PC6,"PC7" : PC7,"PC8" : PC8})
#df3 = pd.DataFrame({ '' : twp1, "PC1" : PC1, "PC2" : PC2,"PC3" : PC3 })
df3.head()
```

```python

# Plot variance ratio of the four principal components
variance_ratio = pca_model.explained_variance_ratio_
print(variance_ratio)

var_percent = variance_ratio * 100

## plot the times of type

#objects = ["PC1", "PC2" , "PC3" ]
objects = ["PC1", "PC2" , "PC3","PC4","PC5","PC6","PC7" ,"PC8"]

y_pos = np.arange(len(objects))

 
plt.bar(y_pos, var_percent, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.title("variance percent with PCA")
 
plt.show()
```

```python
# Run k-means

y_pred = KMeans(n_clusters=3,random_state=1).fit_predict(PCA[:,:3])

df3["KMeans"] = y_pred
clusters_one = df3[df3["KMeans"] == 0] 
clusters_two = df3[df3["KMeans"] == 1] 
clusters_three = df3[df3["KMeans"] == 2] 




 


print(str(np.array(clusters_one[""])) + " is belong to cluster one ")
print("")
print(str(np.array(clusters_two[""]))+ " is belong to cluster two ")
print("")
print(str(np.array(clusters_three[""]))+ " is belong to cluster three ")



# Plot Principal Components 1 and 2, colored by cluster
from matplotlib.colors import ListedColormap
# Create color maps
cmap = ListedColormap(['#ff0000','#fdff00','#2200ff'])
plt.scatter(PCA[:, 0], PCA[:, 1],c=y_pred,  marker="o", cmap=cmap)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()


# Plot Principal Components 1 and 2, colored by cluster
from matplotlib.colors import ListedColormap
# Create color maps
cmap = ListedColormap(['#ff0000','#fdff00','#2200ff'])


fig, ax = plt.subplots()
ax.scatter(PCA[:, 0], PCA[:, 1],c=y_pred,  marker="o", cmap=cmap)

for i, txt in enumerate(zip__):
    ax.annotate(txt, (PCA[:, 0][i],PCA[:, 1][i]))
    
    
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
review = pd.read_csv('./yelp_dataset/review.csv')
```

```python

review = review[['business_id','cool','funny','stars','text','useful']]


#from nltk.corpus import stopwords
#stop = stopwords.words('english')
#review['text'].apply(lambda x: [item for item in x if item not in stop])
#review.to_csv('123.csv')
```

```python

restaurants_r = review[review.business_id.isin(restaurants.business_id.tolist()) == True]
restaurants_r['code'] = 0

shopping_r = review[review.business_id.isin(shopping.business_id.tolist()) == True]
shopping_r['code'] = 1

nightlife_r = review[review.business_id.isin(nightlife.business_id.tolist()) == True]
nightlife_r['code'] = 2

arts_r = review[review.business_id.isin(arts.business_id.tolist()) == True]
arts_r['code'] = 3


business_man = restaurants_r.append(shopping_r).append(nightlife_r).append(arts_r)
```

```python
restaurants_text = " ".join(str(x) for x in restaurants_r[restaurants_r.stars == 5.0].text.tolist())
shopping_text = " ".join(str(x) for x in shopping_r[shopping_r.stars == 5.0].text.tolist())
nightlife_text = " ".join(str(x) for x in nightlife_r[nightlife_r.stars == 5.0].text.tolist())
arts_text = " ".join(str(x) for x in arts_r[arts_r.stars == 5.0].text.tolist())

```

```python
import nltk
from nltk.corpus import stopwords


wordcloud = WordCloud(background_color="white",width=1000, height=860, margin=2,stopwords=set(stopwords.words('english'))).generate(restaurants_text)
wordcloud.to_file('./wordcloud/restaurants_text.png')

wordcloud = WordCloud(background_color="white",width=1000, height=860, margin=2,stopwords=set(stopwords.words('english'))).generate(shopping_text)
wordcloud.to_file('./wordcloud/shopping_text.png')

wordcloud = WordCloud(background_color="white",width=1000, height=860, margin=2,stopwords=set(stopwords.words('english'))).generate(nightlife_text)
wordcloud.to_file('./wordcloud/nightlife_text.png')

wordcloud = WordCloud(background_color="white",width=1000, height=860, margin=2,stopwords=set(stopwords.words('english'))).generate(arts_text)
wordcloud.to_file('./wordcloud/arts_text.png')
```

```python
ml = visitor[['latitude','longitude','review_count','stars','code']]
ml = ml[ml.code !=4]
ml = ml[ml.code !=3]
ml = ml[ml.code !=2]


y = ml.values[:,4]
X = scale(ml.values[:,:3])

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=40)




from sklearn import tree, svm, metrics
from sklearn.metrics import accuracy_score

decisionTree = tree.DecisionTreeClassifier()
decisionTree.fit(X_train, y_train)
pred = decisionTree.predict(X_test)
print("The decision Tree accuracy: " + str(accuracy_score(y_test, pred)))


from sklearn.svm import SVC

svm =  SVC(gamma='auto',C=4)
svm.fit(X_train, y_train)
pred = svm.predict(X_test)
print("The svm accuracy: " + str(accuracy_score(y_test, pred)))

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=0, solver='lbfgs').fit(X_train, y_train)
pred = lr.predict(X_test)
print("The Logistic Regression accuracy: " + str(accuracy_score(y_test, pred)))

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, max_depth=10,random_state=0).fit(X_train, y_train)
pred = rf.predict(X_test)
print("The Random Forest accuracy: " + str(accuracy_score(y_test, pred)))
```
